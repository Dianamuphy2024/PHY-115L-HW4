```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy
```

```python
def f(x):
    return sin(x)
```

```python
x = np.arange(0,2*pi,0.1)
y = f(x)
```

```python
plt.plot(x,y)
```

```python
def dev(f,x,h):
    dev = (f(x+h) - 2*f(x) + f(x-h))/h**2
    return dev
```

```python
x_values = np.arange(0,2*pi,0.1)
devs = np.array([])
for x in x_values:
    devs = np.append(devs,dev(f,x,h))
```

```python
plt.plot(x_values,devs)
```
